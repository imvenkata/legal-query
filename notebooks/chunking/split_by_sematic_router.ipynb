#### Use sematic router splitter for chunking

```python
from semantic_router.encoders import OpenAIEncoder
import tiktoken
# encoding = tiktoken.get_encoding("o200k_base")
enc = tiktoken.encoding_for_model("gpt-4o")

encoder = OpenAIEncoder()
```

```python
from semantic_router.splitters import RollingWindowSplitter
from semantic_router.utils.logger import logger

logger.setLevel("WARNING")

splitter = RollingWindowSplitter(
  encoder=encoder,
  min_split_tokens=100,
  max_split_tokens=500,
  window_size=2,
  plot_splits=True,  # Set to true to visualize chunking
  enable_statistics=True   # to print chunking stats
)
```

```python
# Load the documents from the data directory.
documents = load_documents_from_dir("data/content")
```

```python
splites = splitter(documents[0].page_content)
```

#### Use langchain semantic chunker

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings

text_splitter = SemanticChunker(OpenAIEmbeddings())
```

```python
documents = load_documents_from_dir("data/content")

docs = text_splitter.create_documents([documents[0].page_content])
print(docs[0].page_content)
```

```python
chunks = []
for doc in documents:
    doc_chunks = text_splitter.create_documents([doc.page_content])
    chunks.extend(doc_chunks)
```
